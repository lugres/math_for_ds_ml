```python
# Calculating compound interest in Python - monthly
from math import exp

p = 100 # principal amount
r = .20 # interest rate, by year
t = 2.0 # time span, years
n = 12 # periods, number of months for a year

a=p*(1+(r/n))**(n*t)
print(a) # prints 148.69146179463576
```

```python
# changing periods to every minute
p = 100 # principal amount
r = .20 # interest rate, by year
t = 2.0 # time span, years
n = 525_600 # periods, number of months for a year

a=p*(1+(r/n))**(n*t)
print(a) # prints 149.1824584
```

```python
# Calculating continuous interest in Python
from math import exp

p = 100 # principal, starting amount
r = .20 # interest rate, by year
t = 2.0 # time, number of years

a=p*exp(r*t)
print(a) # prints 149.18246976412703
```

```python
# using limit to calculate Euler's number
from sympy import *

n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)

print(result) # E
print(result.evalf()) # 2.71828182845905
```

Derivatives

```python
# using rise-over-run formula to calculate slope
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x) 
    return m

def my_function(x): 
    return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)
print(slope_at_2) # prints 4.000010000000827
```

```python
# Calculating a derivative in SymPy
from sympy import *

# Declare 'x' to SymPy
x = symbols('x')

# Now just use Python syntax to declare function
f=x**2

# Calculate the derivative of the function
dx_f = diff(f) 
print(dx_f) # prints 2*x

# evaluate derivative at x=2
print(dx_f.subs(x, 2)) # prints 4
```

```python
# Calculating Integrals with approximation - Reimann Sums

def approximate_integral(a, b, n, f): 
    delta_x=(b-a)/n
    total_sum = 0
    print(f'delta_x - {delta_x}')
    
    for i in range(1, n + 1): 
        midpoint=0.5 * (2 * a + delta_x * (2*i-1))
        total_sum += f(midpoint) * delta_x
        print(f'Iter: {i},\tMidpoint: {midpoint:3f},\tTotal sum - {total_sum}')
    
    return total_sum # better to multiply on delta_x here, only once

def my_function(x):
    return x**2 + 1

area = approximate_integral(a=0, b=1, n=5, f=my_function)
print(area)
```

```python
# for 3*x**2 + 1, 0-2

def approximate_integral(a, b, n, f): 
    delta_x=(b-a)/n
    total_sum = 0
    print(f'delta_x - {delta_x}')
    
    for i in range(1, n + 1): 
        midpoint=0.5 * (2 * a + delta_x * (2*i-1))
        total_sum += f(midpoint)
        print(f'Iter: {i},\tMidpoint: {midpoint:3f},\tTotal sum - {total_sum}')
    
    return total_sum * delta_x

def my_function(x):
    return 3*x**2 + 1

area = approximate_integral(a=0, b=2, n=10000, f=my_function)
print(area)
```
