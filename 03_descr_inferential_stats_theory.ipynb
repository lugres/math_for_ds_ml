Central Limit Theorem

```python
import random
import matplotlib.pyplot as plt

sample_size = 50
sample_count = 1000

# Central limit theorem, 1000 samples with 31 uniform random numbers b/w 0.0 and 1.0
x_values = [(sum([random.uniform(0.0, 1.0)*10 for i in range(sample_size)]) / \
sample_size)
for _ in range(sample_count)]

# y_values = [1 for _ in range(sample_count)]

fig, axes = plt.subplots(1, figsize=(14,7))

axes.hist(x=x_values, bins=20)
plt.show()
```
